```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
n = 5000
noise = 0.2
import statsmodels.api as sm
import graphviz
```

### Generating the datasets for 3 cases and fitting regression models

1. Chain

```python
a = np.random.randn(n)
b = a + noise*np.random.randn(n)
c = b + noise*np.random.randn(n)
```

```python
fig = plt.figure(figsize = (8,2))
ax = fig.add_subplot(131)
ax.scatter(a,b,s = 1)
ax.set_xlabel('a')
ax.set_ylabel('b')

ax = fig.add_subplot(132)
ax.scatter(b,c,s = 1)
ax.set_xlabel('b')
ax.set_ylabel('c')
ax = fig.add_subplot(133)
ax.scatter(a,c,s = 1)
ax.set_xlabel('a')
ax.set_ylabel('c')
plt.subplots_adjust(wspace=0.3,hspace=0.3)
#plt.suptitle('Chain plots')
plt.show()
```

```python
X = pd.DataFrame(np.vstack([a, b]).T, columns=['A', 'B'])
X = sm.add_constant(X, prepend=True)
model = sm.OLS(c,X)
results = model.fit()
print(results.summary())
```

2. Fork

```python
b = np.random.randn(n)
a = b + noise*np.random.randn(n)
c = b + noise*np.random.randn(n)
```

```python
fig = plt.figure(figsize = (8,2))
ax = fig.add_subplot(131)
ax.scatter(a,b,s = 1)
ax.set_xlabel('a')
ax.set_ylabel('b')
ax = fig.add_subplot(132)
ax.scatter(b,c,s = 1)
ax.set_xlabel('b')
ax.set_ylabel('c')
ax = fig.add_subplot(133)
ax.scatter(a,c,s = 1)
ax.set_xlabel('a')
ax.set_ylabel('c')
plt.subplots_adjust(wspace=0.3,hspace=0.3)
#plt.suptitle('Fork plots')
plt.show()
```

```python
X = pd.DataFrame(np.vstack([a, b]).T, columns=['A', 'B'])
X = sm.add_constant(X, prepend=True)
model = sm.OLS(c,X)
results = model.fit()
print(results.summary())
```

3. Collider

```python
a = np.random.randn(n)
c = np.random.randn(n)
b = a + c + noise*np.random.randn(n)
```

```python
fig = plt.figure(figsize = (8,2))
ax = fig.add_subplot(131)
ax.scatter(a,b,s = 1)
ax.set_xlabel('a')
ax.set_ylabel('b')
ax = fig.add_subplot(132)
ax.scatter(b,c,s = 1)
ax.set_xlabel('b')
ax.set_ylabel('c')
ax = fig.add_subplot(133)
ax.scatter(a,c,s = 1)
ax.set_xlabel('a')
ax.set_ylabel('c')
plt.subplots_adjust(wspace=0.3,hspace=0.3)
#plt.suptitle('Collider plots')
plt.show()
```
