```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
```

## Graph languages in Python
The GML language is supported by the NetworkX and DoWhy Python libraries. More details on GML and its syntax can be found [here](https://bit.ly/GMLDocs).

Another popular graph language is DOT. A dedicated Python library called pydot [Link](https://bit.ly/PyDOTDocs) allows you to easily read, write, and manipulate DOT graphs in pure Python. DOT is used by graphviz and can be used with NetworkX and DoWhy.

```python
sample = """
graph[
  directed 1

  node[
    id 1
    label "A"
  ]
  node[
    id 2
    label "B"
  ]
  node[
    id 3
    label "C"
  ]
  edge[
    id 1
    source 1
    target 2
  ]
  edge[
    id 2
    source 1
    target 3
  ]
  edge[
    id 3
    source 2
    target 3
  ]
]
"""

graph = nx.parse_gml(sample)
```

```python
fig = plt.figure(figsize=(3,3))
nx.draw(graph, with_labels=True)
plt.show()
```

## Plotting with an adjacency matrix

```python
adj = np.array([[0,1,1],[0,0,1],[0,0,0]])
graph = nx.DiGraph(adj)
fig = plt.figure(figsize=(3,3))
nx.draw(graph, with_labels=True)
plt.show()
```
