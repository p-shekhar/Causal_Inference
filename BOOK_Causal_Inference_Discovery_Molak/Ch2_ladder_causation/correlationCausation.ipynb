You might have heard the phrase that correlation is not causation. Thatâ€™s approximately true.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
u_0 = stats.uniform(-2,4)
u_1 = stats.norm(0,1)
```

```python
sample_size = 10000
x = u_0.rvs(size = sample_size)
n = u_1.rvs(size = sample_size)
y = x**2 + 0.2*n
```

```python
plt.scatter(x,y)
plt.show()
```

```python
## pearson correlation
```

```python
r, p = stats.pearsonr(x, y)
print(f'Correlation coefficient: r = {r:.3f}')
print(f'p-value: p = {p:.3f}')
```

CORRELATION is practically 0 !!!. Although there is clear CAUSATION as seen from structural equations.

### Biased sampling can also be a problem

```python
x_sel = x[(x<-1.9) | (x>1.9)]
n = u_1.rvs(size = len(x_sel))
y = x_sel**2 + 0.2*n
```

```python
plt.scatter(x_sel,y)
plt.show()
```

Due to biased sampling, it is impossible for any method to extract the causal relationship !!
