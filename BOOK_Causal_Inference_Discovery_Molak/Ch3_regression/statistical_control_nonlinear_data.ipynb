```python
import numpy as np
import statsmodels.api as sm
import graphviz
```

### Generating data

```python
n = 5000
a = np.random.randn(n)
x = 2 * a + .7 * np.random.randn(n)
y = 2 * a + 3 * x + .75 * x**2
```

### Fitting the model

```python
X = sm.add_constant(np.stack([x, x**2, a]).T)
model = sm.OLS(y, X)
fitted_model = model.fit()
print(fitted_model.summary(xname=['const', 'x','x^2', 'a']))
```
