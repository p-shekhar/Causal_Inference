```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
n = 1000
# A hypothetical SCM
T = np.random.uniform(20, 110, n)
Y = T + np.random.uniform(0, 40, n)
C = (T + Y < 100).astype('int')
df = pd.DataFrame(np.stack([T, Y, C]).T, columns=['T','Y','C'])
df.head()
```

```python
# Compare average damage (biased vs unbiased)
plt.figure(figsize=(8, 4))
plt.hist(df[df['C'] == 1]['Y'], label='Came back = 1',
    color='r',alpha=.5)
plt.hist(df[df['C'] == 0]['Y'], label='Came back = 0',
    color='b', alpha=.5, bins=25)
plt.xlabel('$Damage$ $severity$', alpha=.5, fontsize=12)
plt.ylabel('$Frequency$', alpha=.5, fontsize=12)
plt.legend()
plt.show()
```
